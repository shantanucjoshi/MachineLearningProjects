# Encoding the columns

 - From the values we would be Label encoding the 'satisfaction', 'Gender', 'Customer Type', 'Type of Travel'
- Even though 'satisfaction' is a target variable we would be converting it to binary data so that it is easy for machine while model building

```python
# from the values we would be Label encoding the 'satisfaction', 'Gender', 'Customer Type', 'Type of Travel'

lst = ['satisfaction', 'Gender', 'Customer Type', 'Type of Travel']

def label_encode(col):
        air_df[col] = air_df[col].astype('category')
        air_df[col] = air_df[col].cat.codes
        
for i in lst:
    label_encode(i)
```

```python
air_df.head()
```

```python
# Since the Class has more than 2 unique values in it we would be using One hot encoding for that column

air_df = pd.get_dummies(air_df,columns =['Class'])
```

```python
air_df.head()
```

```python
# dropping the column = 'Class_Business' as we drop one column in One Hot encoding

air_df.drop(columns = ['Class_Business'], inplace = True)
```

# Splitting the data into target variable and other independent variables


```python

x = air_df.iloc[:,1:]
x.head()
```

```python
y = air_df[['satisfaction']]
```

```python
y.head()
```

# Feature scaling the data

- Here we are using feature scaling as we have a huge value difference in most of the feature so to bring all the features to same scale in order to avoid one feature suppressing or influencing the other variable, scaling is done

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc_x = sc.fit_transform(x)
pd.DataFrame(sc_x)
```

# Splitting the data into Train and Test data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(sc_x, y , test_size = 0.25, random_state = 101)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

# Building logistic regression model

```python
from sklearn.linear_model import LogisticRegression

logit = LogisticRegression()
logit.fit(x_train,y_train)
```

# Prediction

```python
y_pred_train = logit.predict(x_train)

y_pred_test = logit.predict(x_test)
```

# Evaluate the Logistic regression model

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print("Training accuracy for Logistic regression = ", accuracy_score(y_train,y_pred_train ))
print("********" * 5)
print("Test accuracy for Logistic regression= ", accuracy_score(y_test,y_pred_test))
```

```python
print("Training confusion matrix for Logistic regression")
print(confusion_matrix(y_train,y_pred_train ))
print("********" * 5)
print("Test confusion matrix for Logistic regression")
print(confusion_matrix(y_test,y_pred_test))
```

```python
print("Training classification report for Logistic regression")
print(classification_report(y_train,y_pred_train ))
print("***********" * 5)
print("Test classification report for Logistic regression")
print(classification_report(y_test,y_pred_test))
```

# Cross validation

```python
from sklearn.model_selection import cross_val_score
training_accuracy = cross_val_score(logit, x_train, y_train, cv = 20)
test_accuracy = cross_val_score(logit, x_train, y_train, cv = 20)

print("Training cross validation score for Logistic regression")
print(training_accuracy.mean())
print("*************************" * 5)
print("Test cross validation score for Logistic regression")
print(test_accuracy.mean())
```

# Building KNN model

```python
from sklearn.neighbors import KNeighborsClassifier
error_rate = []

# finding the best k value that we would be using to build KNN model(K value where error is the least)

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(x_train,y_train)
    y_pred_i = knn.predict(x_test)
    error_rate.append(np.mean(y_pred_i != np.array(y_test)))
    
```

```python
np.array(y_test)
```

```python
print(error_rate)
```

```python
# plotting the above error rates against the K values

plt.figure(figsize = (16,16), dpi = 100)

plt.plot(range(1,40), error_rate, color = 'blue', linestyle = 'dashed',marker = 'o', markerfacecolor = 'red', markersize = 10)
plt.title("Error Rate vs K_hyperparam_value")
plt.xlabel("K_HyperParam_Value")
plt.ylabel("Error Rate")
plt.show()
```

- From the above plot graph we can see that when k = 3 there is least error so we choose k = 3 while building the model

```python
knn_model = KNeighborsClassifier(n_neighbors = 3)
knn_model.fit(x_train,y_train)
```

```python
y_pred_train_knn = knn_model.predict(x_train)
y_pred_test_knn = knn_model.predict(x_test)
```

## Evaluating the model
